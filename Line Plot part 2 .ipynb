```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [22, 24, 19, 23, 25, 27, 26]
})


sns.lineplot(data = data , x = 'Days', y = 'Temperature' )
plt.show()
```

```python
data = sns.load_dataset('tips')
# data 
```

```python
sns.lineplot(data = data , x = 'total_bill', y = 'tip', hue = 'sex',style ='sex' ,palette='Accent' )
```

```python
sns.lineplot(data = data , x = 'total_bill', y = 'tip', hue = 'sex',style ='sex' ,palette='hot',dashes=False ,marker = '^', legend='full')
```

```python
sns.set(style ='dark')

sns.lineplot(data = data , x = 'total_bill', y = 'tip', hue = 'sex',style ='sex' ,palette='hot',dashes=False ,marker = '^', legend=False)
# plt.xlabel('')
# plt.ylabel()
plt.title('Line Plot ')
plt.show()
```
