```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = sns.load_dataset("iris")
correlation = data.corr(numeric_only=True)

# Advanced heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation,
    annot=True,               # Show values inside boxes
    fmt=".2f",                # Format for the annotation
    cmap="viridis",          # Colormap
    cbar=True,                # Show color bar
    linewidths=0.5,           # Line width between cells
    linecolor='black',        # Line color
    square=True,              # Square shaped cells
    xticklabels=True,         # Show x-axis labels
    yticklabels=True,         # Show y-axis labels
    mask=None,                # Mask upper/lower triangle if needed
    cbar_kws={"shrink": 0.8, "label": "Correlation"},  # Custom colorbar
    center=0,                 # Center of the colormap
)

plt.title("Advanced Correlation Heatmap", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample dataset
data = {
    'Math_Score':      [78, 85, 96, 70, 65, 88, 92, 55, 60, 80],
    'Science_Score':   [82, 89, 94, 72, 67, 90, 95, 58, 63, 85],
    'English_Score':   [75, 80, 91, 74, 68, 85, 88, 60, 62, 78],
    'Study_Hours':     [4, 6, 8, 3, 2, 7, 9, 1, 2, 5],
    'Attendance':      [90, 95, 98, 85, 80, 96, 99, 70, 75, 93]
}

df = pd.DataFrame(data)
df
```

```python

correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix,
    annot=True,        # Show correlation values
    fmt=".2f",         # Format float values
    cmap="coolwarm",   # Color map
    linewidths=0.5,    # Line width between boxes
    square=True        # Make cells square
)

plt.title("Student Performance Correlation Heatmap")
plt.tight_layout()
plt.show()
```
