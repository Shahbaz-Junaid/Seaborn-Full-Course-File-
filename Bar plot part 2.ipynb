# Bar Plot Part 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python

tips_df = sns.load_dataset('tips')
```

```python
tips_df.head(3)
```

```python
tips_df.columns
```

# Average Total Bill by Day

```python
sns.barplot(x='day', y='total_bill', data=tips_df, estimator='mean')
plt.title("Average Total Bill by Day")
plt.show()
```

# Total Tip by Sex

```python
sns.barplot(x='sex', y='tip', data=tips_df, estimator='sum')
plt.title("Total Tips Collected by Gender")
plt.show()
```

# Average Tip by Smoker Status

```python
sns.barplot(x='smoker', y='tip', data=tips_df, estimator='mean')
plt.title("Average Tip by Smoker vs Non-Smoker")
plt.show()
```

# Total Bill by Time (Lunch vs Dinner)

```python
sns.barplot(x='time', y='total_bill', data=tips_df, estimator='sum')
plt.title("Total Bill Amount by Time of Day")
plt.show()
```

# Average Tip by Group Size

```python
sns.barplot(x='size', y='tip', data=tips_df, estimator='mean')
plt.title("Average Tip by Group Size")
plt.show()
```

# Bill by Gender and Day (Hue Usage)

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df)
plt.title("Total Bill by Gender and Day")
plt.show()
```

# Tips by Smoker and Time

```python
sns.barplot(x='time', y='tip', hue='smoker', data=tips_df)
plt.title("Tips by Smoker Status and Time")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.barplot(
    x='sex', 
    y='tip', 
    hue='day', 
    data=tips_df,
    estimator='mean',           # Use average tip
    ci='sd',                    # Show standard deviation as error bars
    palette='Set2',             # Predefined color palette
    capsize=0.1,                # Small cap on error bars
    width=0.7,                  # Bar width
    order=['Male', 'Female']    # Custom order
)

plt.title('Average Tip by Gender, Split by Day', fontsize=15)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Tip ($)', fontsize=12)
plt.legend(title='Day')
plt.tight_layout()
plt.show()
```
