```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head(3)
```

```python
sns.distplot(tips_df["size"])
# plt.show()
```

```python
# KDE  mean kernal denstiy Estimator 
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
tips_df.columns
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["tip"],bins = 50)
```

```python
sns.distplot(tips_df["total_bill"],bins = 50, hist = False,kde=True)
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde= False,color='red')
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde= False,color='red',vertical=True)
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,axlabel='Totalbill',label='tips')
plt.legend()
```
