# Histogram Using Displot Part 2 

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd 
```

```python
tips_df = sns.load_dataset('tips')
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,axlabel='Totalbill',label='tips')
plt.legend()
```

```python
plt.figure(figsize=(10,9))
sns.set(style = 'dark')
sns.distplot(tips_df["total_bill"],fit=norm,axlabel='Totalbill',label='tips',color = 'red')
plt.legend()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cleaned and sorted bins (removed duplicates and sorted)
bins = sorted(list(set([1, 5, 10, 15, 20, 30, 35, 40, 45, 50, 55])))

sns.set(style='dark')
sns.histplot(tips_df["total_bill"], bins=bins, kde=True)

plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cleaned and sorted bins
bins = sorted(list(set([1, 5, 10, 15, 20, 30, 35, 40, 45, 50, 55])))

sns.set(style='dark')
sns.histplot(
    tips_df["total_bill"],
    bins=bins,
    color='skyblue',
    edgecolor='black',
    linewidth=1.5,
    alpha = 0.7
)

plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```
