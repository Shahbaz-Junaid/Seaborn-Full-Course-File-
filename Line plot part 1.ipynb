# Load the DataSet using Seaborn and its function name is load_dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("flights")
# data
# Line plot
sns.lineplot(data=data, x="year", y="passengers", hue="month")
plt.title("Monthly Flight Passengers Over the Years")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample employee salary data
df = pd.DataFrame({
    'Year': [2020, 2021, 2022, 2023, 2024],
    'IT': [55000, 58000, 60000, 62000, 65000],
    'HR': [40000, 42000, 45000, 47000, 50000]
})

# Melt to long format for seaborn
df_melted = df.melt('Year', var_name='Department', value_name='Salary')

# Line plot
sns.lineplot(data=df_melted, x='Year', y='Salary', hue='Department', marker='o')
plt.title('Salary Growth by Department')
plt.grid(True)
plt.show()
```
