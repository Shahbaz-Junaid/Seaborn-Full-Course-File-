# Scatter Plot Using the Seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips,palette='hot',hue = 'sex',markers='s')

# Add plot title
plt.title("Scatter Plot of Total Bill vs Tip")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the sample dataset
tips = sns.load_dataset("tips")

# Create an advanced scatter plot
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="sex",              # Color by gender
    style="smoker",         # Different marker style for smokers/non-smokers
    size="size",            # Size of points based on party size
    palette="Set2",         # Use a pleasant color palette
    alpha=0.8               # Slight transparency
)

# Add title and labels
plt.title("Advanced Scatter Plot: Total Bill vs Tip", fontsize=14)
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Show legend and plot
plt.legend(title="Legend", bbox_to_anchor=(1.05, 1), loc=2)
plt.bar_label()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
tips = sns.load_dataset("tips")

# Scatter plot with color scale based on 'size' of the party
scatter = plt.scatter(
    x=tips["total_bill"],
    y=tips["tip"],
    c=tips["size"],          # Color based on party size
    cmap='viridis',          # Colormap
    s=100,                   # Size of points
    alpha=0.8,               # Transparency
    edgecolors='w'           # White border
)

# Add color bar (scale)
cbar = plt.colorbar(scatter)
cbar.set_label('Party Size')

# Labels and title
plt.title("Scatter Plot with Color Scale")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Drop missing values
titanic = titanic.dropna(subset=["age", "fare", "survived", "sex"])
plt.figure(figsize=(18,9))

# Create scatter plot
sns.scatterplot(
    data=titanic,
    x="age",
    y="fare",
    hue="survived",      # Color by survival
    style="who",         # Marker shape by gender
    palette="Set1",
    s=90,                # Marker size
    alpha=0.7
)

# Add title and labels
plt.title("Titanic: Age vs Fare (Style by Sex, Hue by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived & Sex", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Clean data: Drop missing values
titanic = titanic.dropna(subset=["age", "fare", "survived", "sex", "pclass"])

# Create the advanced scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=titanic,
    x="age",
    y="fare",
    hue="survived",         # Color by survival
    style="who",            # Marker shape by gender
    size="pclass",          # Size by class (1st > 2nd > 3rd)
    sizes=(50, 200),        # Size scale
    palette="coolwarm",     # Color palette
    alpha=0.8,
    edgecolor="black",
    linewidth=0.5
)

# Add plot enhancements
plt.title("Titanic Advanced Scatter Plot: Age vs Fare", fontsize=14, weight='bold')
plt.xlabel("Passenger Age")
plt.ylabel("Fare Paid")
plt.grid(True, linestyle="--", alpha=0.5)

# Move legend outside
plt.legend(title="Survival / Sex / Class", bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()

# Show plot
plt.show()
```
